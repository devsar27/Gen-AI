```python
import tiktoken
```

```python
tokenizer = tiktoken.encoding_for_model(model_name="gpt-4")
```

```python
tokenId = tokenizer.encode(text)
tokenId
```

```python
resultIds = [47,76,315,6890,374,35489,41,72,13]
```

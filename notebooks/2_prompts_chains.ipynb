## Chain - Runnable  
What is a Runnable?

A Runnable is anything that can take an input and produce an output.

Think of it like a Python function.

input ---> Runnable ---> output

For example,

Question
    │
    ▼
ChatGoogleGenerativeAI
    │
    ▼
Answer

Here, the LLM itself is a Runnable.

```python
# res = llm.invoke(prompts.invoke({"language": "Hinghlish", "query": "I love to play cricket and football."}))
# res.content
```

```python
chains = prompts | llm | out | transform_case
```

```python
res = chains.invoke({"language": "Hinglish", "query": "I love Langchain"})
print(res)  
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int, b:int):
    """
      It will return the sum of two numbers 

      Args:
         a : N1 
         b : N2 
    """
    return a + b 

@tool
def multiply_numbers(a:int, b:int):
    """
      It will return the sum of two numbers 

      Args:
         a : N1 
         b : N2 
    """
    return a * b 
```

```python
add_numbers.invoke({"a": 12 , "b": 3})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
agent = create_agent(
    model = llm, 
    tools = [add_numbers, multiply_numbers],
    system_prompt = "You are a match teacher, and always use tool for calculation."
)
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "What is 2 + 30  * 10 ?"}]})
```

```python
print(response['messages'][-1].content)
```

```python
from dotenv import load_dotenv
load_dotenv() 
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm  = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts = [
      {"role": "user", "content": "Hello, My name is Dev Saraf."},
    {"role": "ai", "content": "Hello Dev Saraf! It's nice to meet you. How can I help you today?"},
    {"role": "user", "content": "What is my name?"}
]
res = llm.invoke(prompts)
```

## Manage History 

```python
history = []  ## [{"role": "user", "content": "WHo is the CEO of Google?"}]
while True:
    query = input("User :")
    if query.lower() in ["exit", "quit", "bye"]:
        print("Good bye! 👋")
        break
    history.append( {"role": "user", "content": query})
    print("User :", query)
    res = llm.invoke(history)
    history.append({"role": "ai", "content": res.content})
    print("AI :", res.content, "\n\n")


```
